```python
import pandas as pd
data = pd.read_csv('Train.csv')
data.head(2)
```

```python
# preprocessing
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from string import punctuation
stopword_list = stopwords.words('english')
def preprocess(data):
    clean_text = [word.lower() for word in word_tokenize(data) if (word not in punctuation) and (word.lower() not in stopword_list) and (word.isalpha())]
    return clean_text
```

```python
final_text = data.text.apply(preprocess)
final_text
```

```python
def filter_1(data):
    clean_text = [word for word in data if (len(word)>2)]
    return clean_text
```

```python
final_text = final_text.apply(filter_1)
final_text
```

```python
final_data = final_text.to_list()
final_data
```

```python
from gensim.models import Word2Vec
model = Word2Vec(final_data,window=6,min_count=2,epochs=2,vector_size=200)
```
